```python
import numpy as npfrom Rmagic import rmagicrmagic.shell = {'X': np.random.standard_normal(40),                'Y': np.random.standard_normal(40)}
```

```python
snippet = '''a=lm(Y~X)# since I called print, I will see the summary of the modelprint(summary(a))# this will produce one plotplot(X, Y, pch=23, bg='orange', cex=2)# by default R creates 4 plots for plotting an lm, so there will be five plots in totalplot(a)# this next print statement will show me summary(X)print(summary(X))r = resid(a)'''
```

```python
opts = '--inputs=X,Y --outputs=r --bg="gray" --width=700'
```

```python
rmagic.cell_magic(opts, snippet)
```

```python
rmagic.shell['r']
```
